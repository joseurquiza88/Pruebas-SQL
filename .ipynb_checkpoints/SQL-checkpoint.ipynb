```python
# ----- Where ----- 
# Objetivo: Crear un df de jugadores que esten valuados por mas de $100 mil EU y que sean de Brasil
q = """ SELECT * FROM jugadores 
WHERE country_of_birth = 'Brazil' AND 
market_value_in_eur >= 100000 """

df_ejercicio = pysqldf (q)
df_ejercicio

### Setamos algunas columnas nomas
qb = """ SELECT last_name, country_of_birth, market_value_in_eur FROM jugadores 
WHERE country_of_birth = 'Brazil' AND 
market_value_in_eur >= 100000 """

df_ejercicio = pysqldf (qb)
df_ejercicio
```

## Operaciones mas avanzadas
### Group by

```python
jugadores.columns 
q = """
SELECT current_club_name AS club, 
AVG(highest_market_value_in_eur) AS promedio_valor
FROM jugadores
GROUP BY 1
ORDER BY promedio_valor DESC
LIMIT 11
"""

top_10_clubes= pysqldf(q)
top_10_clubes


####
```

### HAVING
Misma tabla pero teniendo en cuenta solo clubes con valor promeio menor a 15 millones EU
Se aplica el HAVING y no el WHERE ya que el friltro se aplica sobre el valor ya agrupado.

```python
jugadores.columns 
q = """
SELECT current_club_name AS nombres_del_club, 
AVG(highest_market_value_in_eur) AS promedio_valor
FROM jugadores
GROUP BY 1
HAVING promedio_valor < 15000000
ORDER BY promedio_valor DESC
LIMIT 11
"""

top_10_clubes= pysqldf(q)
top_10_clubes


####
```

### LEFT JOIN

```python
#La llave de ambos df son: a.nombres_del_club = b.name
q = """
SELECT a.*,
b.stadium_name
FROM top_10_clubes A
LEFT JOIN clubes b
ON a.nombres_del_club = b.name
"""
pysqldf(q)
```
